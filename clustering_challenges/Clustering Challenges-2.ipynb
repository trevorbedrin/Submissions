```python
%matplotlib inline
import matplotlib.pyplot as plt
from pymongo import MongoClient
from textblob import TextBlob
from sklearn.cluster import KMeans
from sklearn.cluster import MiniBatchKMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
from pprint import pprint
from sklearn import preprocessing
from sklearn.feature_extraction.text import TfidfVectorizer
from collections import defaultdict
from sklearn.metrics.pairwise import pairwise_distances_argmin_min
import numpy as np
import pandas as pd
```

```python
client = MongoClient()
db = client.dsbc
db.collection_names()
survey_results = db.survey_results
survey_comments = db.survey_comments
```

```python
negative_data = []
neutral_data = []
positive_data = []
polar_data = []

responses = survey_comments.find({'type':'msct workshop 1'})
print responses.count()
for response in responses:
    response_text = response['response']
    blob = TextBlob(response_text)
    response_code = response['code']
    response_type = response['type']
    response_sentiment = float(response['sentiment'])
    if response_sentiment < 0:
        negative_data.extend(blob.sentences)
        polar_data.extend(blob.sentences)
    elif response_sentiment > 0:
        positive_data.extend(blob.sentences)
        polar_data.extend(blob.sentences)
    else:
        neutral_data.extend(blob.sentences)
```

```python
data = [sentence.string for sentence in negative_data]
#data = [sentence.string for sentence in positive_data]
#data = [sentence.string for sentence in polar_data]

vectorizer = TfidfVectorizer(ngram_range=(1,3), max_features=10000, stop_words='english')
X = vectorizer.fit_transform(data)
print X.shape
```

## Challenge 8

Download this dataset on the purchase stats from clients of a wholesale distributor. Cluster the clients based on their annual spending features (fresh, milk, grocery, frozen, detergents_paper, delicatessen). Remember to scale the features before clustering. After finding a reasonable amount of clusters, for EACH cluster, plot the histogram for every single feature: FRESH, MILK, GROCERY, FROZEN, DETERGENTS_PAPER, DELICATESSEN, CHANNEL, REGION. Is there a natural way to characterize each cluster? How would you describe each cluster to the wholesale distributor if you were working for them?

```python
dataframe = pd.read_csv('Wholesale customers data.csv', dtype=float)
X = dataframe[["Fresh","Milk","Grocery","Frozen","Detergents_Paper","Delicassen"]]
X_dict = X.to_dict()
```

```python
scaled_X = preprocessing.scale(X)
```

```python
model = AgglomerativeClustering(n_clusters=16, linkage='average', affinity='euclidean')

clusters = model.fit_predict(scaled_X)
cluster_contents = defaultdict(list)

for i in range(0, len(dataframe)):
    cluster_contents[clusters[i]].append([X_dict['Fresh'][i], X_dict['Milk'][i], X_dict['Grocery'][i], 
                                          X_dict['Frozen'][i], X_dict['Detergents_Paper'][i], X_dict['Delicassen'][i]])

for i in range(0, len(cluster_contents)):
    cluster_df = pd.DataFrame(cluster_contents[i], columns=['Fresh','Milk','Grocery','Frozen','Detergents_Paper','Delicassen'])
    plt.figure(figsize=[15,15])
    plt.subplot(len(cluster_contents), 6, (i*6)+1)
    plt.hist(cluster_df["Fresh"], bins=10, range=[0,50000])
    plt.subplot(len(cluster_contents), 6, (i*6)+2)
    plt.hist(cluster_df["Milk"], bins=10, range=[0,50000])
    plt.subplot(len(cluster_contents), 6, (i*6)+3)
    plt.hist(cluster_df["Grocery"], bins=10, range=[0,50000])
    plt.subplot(len(cluster_contents), 6, (i*6)+4)
    plt.hist(cluster_df["Frozen"], bins=10, range=[0,50000])
    plt.subplot(len(cluster_contents), 6, (i*6)+5)
    plt.hist(cluster_df["Detergents_Paper"], bins=10, range=[0,50000])
    plt.subplot(len(cluster_contents), 6, (i*6)+6)
    plt.hist(cluster_df["Delicassen"], bins=10, range=[0,50000])
 
```
