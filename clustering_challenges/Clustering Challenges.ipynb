```python
%matplotlib inline
import matplotlib.pyplot as plt
from pymongo import MongoClient
from textblob import TextBlob
from sklearn.cluster import KMeans
from sklearn.cluster import MiniBatchKMeans
from pprint import pprint
from sklearn import preprocessing
from sklearn.feature_extraction.text import TfidfVectorizer
from collections import defaultdict
from sklearn.metrics.pairwise import pairwise_distances_argmin_min
import numpy as np
import pandas as pd
```

```python
client = MongoClient()
db = client.dsbc
db.collection_names()
survey_results = db.survey_results
survey_comments = db.survey_comments
```

```python
negative_data = []
neutral_data = []
positive_data = []
polar_data = []

responses = survey_comments.find({'type':'msct workshop 1'})
print responses.count()
for response in responses:
    response_text = response['response']
    blob = TextBlob(response_text)
    response_code = response['code']
    response_type = response['type']
    response_sentiment = float(response['sentiment'])
    if response_sentiment < 0:
        negative_data.extend(blob.sentences)
        polar_data.extend(blob.sentences)
    elif response_sentiment > 0:
        positive_data.extend(blob.sentences)
        polar_data.extend(blob.sentences)
    else:
        neutral_data.extend(blob.sentences)
```

```python
data = [sentence.string for sentence in negative_data]
#data = [sentence.string for sentence in positive_data]
#data = [sentence.string for sentence in polar_data]

vectorizer = TfidfVectorizer(ngram_range=(1,3), max_features=10000, stop_words='english')
X = vectorizer.fit_transform(data)
print X.shape
```

## Challenge 1

Cluster sentences with K-means. If you have your own Fletcher test data, get sentences out and cluster them. If not, cluster the tweets you gathered during the Twitter API challenge.
```python
def cluster_comments(k=8, X=None):
    k_means_model = KMeans(n_clusters=k)
    clusters = k_means_model.fit_predict(X)

    cluster_contents = defaultdict(list)

    for i in range(0, len(data)):
        cluster_contents[clusters[i]].append(data[i])

    for i in range(0,k):    
        print "Cluster #", i+1
        pprint (cluster_contents[i])
        print

    return (k_means_model, clusters, cluster_contents)

used_k = 16        
(k_means_model, clusters, cluster_contents) = cluster_comments(used_k, X)
```

Instead of using EVERY single point to find the new place of the centroid, MiniBatch just randomly samples a small number (like 100) in the cluster to calculate the new center.
