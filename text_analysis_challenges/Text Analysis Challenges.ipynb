## Challenge 1

Calculate the tf-idf for each word in the first 100 reviews in the nltk movie reviews corpus. For each document, print the top 10 tf-idf words.

```python
import nltk
from nltk.corpus import movie_reviews
from nltk.corpus import stopwords
from collections import defaultdict
import re

tf_idf = []

stop = stopwords.words('english')
stop.extend(['.', ',', '(', ')', "'", '"',"-", "--", "?", ";", ":", " ", "!", "#", "%", "$", "'", "&"])

fileids = movie_reviews.fileids()[:10]
doc_words = [movie_reviews.words(fileid) for fileid in fileids]
documents = [' '.join(words) for words in doc_words]
dictionary = ' '.join(documents)

dictionary_tokens = re.split('\s+', dictionary)
dictionary_tokens = [i for i in dictionary_tokens if i not in stop]

dictionary_frequency = defaultdict(int)
doc_frequency_list = []

for word in dictionary_tokens:
    dictionary_frequency[word] += 1
for k, v in dictionary_frequency.iteritems():
    dictionary_frequency[k] = float(v)/len(dictionary_tokens)
    
for doc in documents:
    doc_frequency_tuples = []
    doc_frequency = defaultdict(int)
    doc_tokens = re.split('\s+', doc)
    doc_tokens = [i for i in doc_tokens if i not in stop]
    for word in doc_tokens:
        doc_frequency[word] += 1
    for k, v in doc_frequency.iteritems():
        doc_frequency[k] = float(v)/len(doc_tokens)
        temp = float(doc_frequency[k])/dictionary_frequency[k]
        doc_frequency_tuples.append((temp, k))
    tf_idf.append(sorted(doc_frequency_tuples,reverse=True))


for index, thingy in enumerate(tf_idf):
    print "Document ", index + 1, ": "
    for i, j in thingy[0:10]:
        print "  ", j, ": " , i
    print
```

## Challenge 2

Calculate the tf-idf for each word in the tweets you uploaded to your mongodb. For each tweet, print the highest tf-idf term. What are potential problems when trying to calculate tf-idf on really short documents like tweets?

```python
from pymongo import MongoClient
from textblob import TextBlob
```

```python
client = MongoClient()
db = client.dsbc
starcitizen = db.starcitizen
```

```python
tweets = starcitizen.find({})
tf = []
dictionary_dict = defaultdict(int)

tf_idf = []

for index, tweet in enumerate(tweets):
    tf_dict = defaultdict(dict)
    blob = TextBlob(tweet['tweet'])
    for word in blob.words:
        dictionary_dict[word] += 1
        tf_dict[word] = float(blob.words.count(word)) / len(blob.words)
    tf.append(tf_dict)

for item in tf:
    tuple_list = [] 
    for k, v in item.iteritems():
        temp = float(v) / dictionary_dict[k]
        tuple_list.append((temp, k))    
    tf_idf.append(sorted(tuple_list,reverse=True))

```

```python
for index, thingy in enumerate(tf_idf):
    print "Tweet ", index + 1, ": "
    for i, j in thingy[0:10]:
        print "  ", j, ": " , i
    print
```

One potential problem is that, because of the brevity of the tweets, there are equal numbers of every word.  Thic would cause there to be equal tf-idf values for every word in the tweet.
