```python
import pickle
```

```python
with open('database_file', 'r') as database_file:
    database = pickle.load(database_file)
```

```python
import pandas as pd

df = pd.DataFrame(database)
df.columns = pd.Index([u'director', u'domestic_total_gross', u'movie_title', u'rating', u'release_date', u'runtime'])
```

```python
df1 = df.dropna(how='any')
```

```python
df1
```

### Pandas Challenge #1

```python
%matplotlib inline
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(16,8))
df2 = df1.sort(columns=['release_date'])
df2.plot(x='release_date', y='domestic_total_gross')
```

### Pandas Challence #2

```python
df3 = df1[df1.runtime < 500]
df3.plot(x='domestic_total_gross', y='runtime', kind='scatter')
```

### Pandas Challenge #3

```python
rating_df = df1.groupby('rating')

rating_mean = rating_df.mean().T
rating_mean
```

### Pandas Challenge #4

```python
fig = plt.figure(figsize=(16,16))
counter = 1
for letter, grp in rating_df:
    plt.subplot(12,1,counter)
    counter += 1
    grp.plot(x='release_date', y='domestic_total_gross').set_title(letter)
```

### Pandas Challenge #5

```python
df4 = df1.sort(columns=['director'])[5:]

director_df = df4.groupby('director').mean().sort(columns=['domestic_total_gross'], ascending=False)
director_df[0:1]
```

### Pandas Challenge #6

```python
axes = df1['release_date'].map(lambda x: x.year)
df5 = df1
df5['year'] = axes
#df5.head()

year_df = df5.groupby('year')
year_df.size()
```

```python
fig = plt.figure(figsize=(16,8))
std_dev = year_df.std()['domestic_total_gross']
year_df.mean()['domestic_total_gross'].plot(kind='bar', yerr=std_dev)
```
