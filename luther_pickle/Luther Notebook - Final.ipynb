# Design draft - Project Luther

## Client

<b>Fill In The Gap Productions</b>, an up and coming movie production company, is looking to identify the best potential projects to pursue to maximize revenue over the course of a calendar year.  They have approached us to test whether their business model, and its assumptions, are valid.




## Client's question/need:

The client's business model is to identify the best possible projects to pursue that will yield the highest possible total gross domestic and opening weekend ticket sales using historical movie release data.  Their hypothesis is that the release month, genre, rating, and number of similar projects being released that month are the best predictors of how well a project will do and want us to test this. 

## My "population" of movies:

The population of movies used for this project will be <b>all</b> movies currently listed on <a href="www.boxofficemojo.com">Box Office Mojo</a> that have budget information.  Currently this includes approximately 2,755 movie titles.

## Features per movie:

We will need to collect the data points the client wishes to test:
    <ul>
        <li>Release Month</li>
        <li>Genre</li>
        <li>MPAA Rating</li>
    </ul>
    
We will also capture the following information to test whether the client's target data points are the best to use:
    <ul>
        <li>Director(s)</li>
        <li>Featured Cast</li>
        <li>Producer(s)</li>
        <li>Production Budget</li>
        <li>Runtime</li>
    </ul>    


```python
# Imports and function defs
%matplotlib inline
import matplotlib.pyplot as plt
import pickle
import pandas as pd
import numpy as np
from itertools import groupby
from collections import defaultdict
import operator
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from pprint import pprint

def train_model(dataframe, explanatory_vars, outcome_var, alpha):
    X = dataframe[explanatory_vars]
    Y = dataframe[outcome_var]
    model = linear_model.Ridge(alpha)
    model.fit(X,Y)    
    return model

def plot_predictions(dataframe, explanatory_vars, outcome_var, model):
    X = dataframe[explanatory_vars]
    # plot the predictions 
    plt.plot(dataframe[outcome_var], model.predict(X), 'go')
    # plot the 1:1 actual outcome line
    plt.plot(dataframe[outcome_var], dataframe[outcome_var], 'b')
    # axis labels
    plt.xlabel('Actual %s' % outcome_var)
    plt.ylabel('Predicted %s' % outcome_var)
    
def plot_exp_of_predictions(dataframe, explanatory_vars, outcome_var, model):
    X = dataframe[explanatory_vars]
    log_predictions = model.predict(X)
    # Take the exp of these log predictions to get predictions!
    plt.plot(np.exp(dataframe[outcome_var]), np.exp(log_predictions), 'go')
    plt.plot(np.exp(dataframe[outcome_var]), np.exp(dataframe[outcome_var]), 'b')
    plt.xlabel('Actual Exp(%s)' % outcome_var)
    plt.ylabel('Predicted Exp(%s)' % outcome_var)
    
def get_top_items(dataframe, column, num_items):
    item_list = [item for item in dataframe[column]]

    item_dict = defaultdict(int)
    for item in item_list:
        item_dict[item] += 1
    
    item_dataframe = pd.DataFrame(item_dict.items(), columns=[column, 'Count'])
    item_percentage_series = pd.Series([float(item)/len(item_list) for k, item in item_dict.items()], 
                                       name='PercentageOfTotal')
    item_dataframe = pd.concat([item_dataframe, item_percentage_series], axis=1)

    item_list = [item for item in item_dataframe.sort('PercentageOfTotal', 
                                                      ascending=False)[0:num_items][column]]
    return item_list

def find_best_polynomial(trainingframe, testingframe, y, feature, degree, alpha=0):
    ''' find the degree of the polynomial that minimizes trainign error using dependent variable x
    to predict y '''
    training_error = []
    X_training = trainingframe[[y,feature,'Ones']]
    X_testing = testingframe[[y,feature,'Ones']]
    X_training = X_training.dropna(how='any')
    X_testing = X_testing.dropna(how='any')
    param_list = ['Ones']
    
    for i in range(0,degree+1):
        if i > 0:
            X_training['feature%d' % i] = [np.power(item, i) for item in X_training[feature]]
            X_testing['feature%d' % i] = [np.power(item, i) for item in X_testing[feature]]
            param_list.append('feature%d' % i)
        model = train_model(X_training, param_list, y, alpha)
        training_error.append((mean_squared_error(X_training[y],model.predict(X_training[param_list])),mean_squared_error(X_testing[y],model.predict(X_testing[param_list]))))

    return training_error
```

```python
# Load Pickle file and build full DataFrame object
with open('database_file', 'r') as database_file:
    database = pickle.load(database_file)

full_data_frame = pd.DataFrame(database)
full_data_frame.columns = pd.Index([u'Actors', u'Budget', u'Composers', u'Directors', 
                                    u'DomesticTotalGross', u'Genre', u'Title', u'OpeningWeekend', 
                                    u'Producers', u'Rating', u'ReleaseDate', u'Runtime'])
```

```python
# Process full DataFrame into workable subset
index_series = full_data_frame.index.values
working_data_frame = pd.DataFrame(full_data_frame[['DomesticTotalGross'
                                                   ,'Budget'
                                                   ,'ReleaseDate'
                                                   ,'Runtime'
                                                   ,'Genre'
                                                   ,'Rating'
                                                   ]])
working_data_frame = working_data_frame.dropna(how='any')
working_data_frame['Budget'] = [float(item) for item in working_data_frame['Budget']]
working_data_frame['BudgetLog'] = [np.log(float(item)) for item in working_data_frame['Budget']]
working_data_frame['BudgetSquared'] = [np.power(float(item),2) for item in working_data_frame['Budget']]
working_data_frame['BudgetCubed'] = [np.power(float(item),3) for item in working_data_frame['Budget']]
working_data_frame['RuntimeSquared'] = [np.power(float(item),2) for item in working_data_frame['Runtime']]
working_data_frame['RuntimeCubed'] = [np.power(float(item),3) for item in working_data_frame['Runtime']]
working_data_frame['Ones'] = 1.0
working_data_frame['LogGross'] = [np.log(item) for item in working_data_frame['DomesticTotalGross']]
working_data_frame['ReleaseMonth'] = [item.strftime('%b') for item in working_data_frame['ReleaseDate']]
working_data_frame['LogMonth'] = [np.log(item.month) for item in working_data_frame['ReleaseDate']]
working_data_frame['ReleaseYear'] = [item.year for item in working_data_frame['ReleaseDate']]
working_data_frame = pd.concat([working_data_frame,pd.get_dummies(working_data_frame['ReleaseMonth'])], 
                               axis=1)
working_data_frame = pd.concat([working_data_frame,pd.get_dummies(working_data_frame['Rating'])], 
                               axis=1)

# Build and add genre/rating features and intercept feature
genre_list = [x for x in reduce(operator.add, [item.split(' ') for item 
                                               in working_data_frame['Genre']]) if x != '/']
genre_dict = defaultdict(int)
for item in genre_list:
    genre_dict[item] += 1
genre_list = [key for key, item in genre_dict.iteritems()]
for item in genre_list:
    working_data_frame[item] = 0.0

for k, v in working_data_frame['Genre'].iteritems():
    for item in working_data_frame.loc[[k]]['Genre'].values:
        for genre_item in item.split(' '):
            if genre_item != '/':
                working_data_frame.loc[k, genre_item] = 1

# Limit by year
#year_list = range(1990, 2014)
#working_data_frame = working_data_frame[working_data_frame['ReleaseYear'].isin(year_list)]
#working_data_frame
#len(working_data_frame)
```

```python
# Generate interaction features
rating_list = ['R','PG','PG-13','G','Unrated','NC-17','GP','M','X']
month_list = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

rating_genre_interaction_list = []
for rating in rating_list:
    for genre in genre_list:
        working_data_frame['%s_%s' % (rating, genre)] = working_data_frame[rating] * working_data_frame[genre]
        rating_genre_interaction_list.append('%s_%s' % (rating, genre))

month_genre_interaction_list = []        
for month in month_list:
    for genre in genre_list:
        working_data_frame['%s_%s' % (month, genre)] = working_data_frame[month] * working_data_frame[genre]
        month_genre_interaction_list.append('%s_%s' % (month, genre))
```

```python
# Set features to use
x_parameter_list = ['Ones']
#x_parameter_list.append('ReleaseMonth')
#x_parameter_list.append('LogMonth')
#x_parameter_list.append('BudgetLog')
#x_parameter_list.append('Budget')
#x_parameter_list.append('BudgetSquared')
#x_parameter_list.append('BudgetCubed')
x_parameter_list.append('Runtime')
#x_parameter_list.append('RuntimeSquared')
#x_parameter_list.append('RuntimeCubed')
x_parameter_list.extend(genre_list)
x_parameter_list.extend(get_top_items(working_data_frame, 'Rating', 1000))
x_parameter_list.extend(get_top_items(working_data_frame, 'ReleaseMonth', 12))
x_parameter_list.extend(rating_genre_interaction_list)
x_parameter_list.extend(month_genre_interaction_list)

len(x_parameter_list)
```

```python
# Randomize Data and split into training and testing sets
randomized_data_frame = working_data_frame.reindex(np.random.permutation(working_data_frame.index))
dataframe_size = int(len(randomized_data_frame.index)*.75)
randomized_training_set = randomized_data_frame[:dataframe_size]
randomized_testing_set = randomized_data_frame[dataframe_size:]
#randomized_testing_set.dtypes
```

```python
alphas = [i*.01 for i in range(1,10001)]
```

```python
# Build models with varying alphas
lin_coefs = []
lin_errors = []
for i in alphas:
    model = train_model(randomized_training_set, x_parameter_list, 'DomesticTotalGross', i)
    lin_coefs.append(model.coef_)
    lin_errors.append(mean_squared_error(randomized_testing_set['DomesticTotalGross'],
                       model.predict(randomized_testing_set[x_parameter_list])))
#print lin_coefs[lin_errors.index(min(lin_errors))]   
print alphas[lin_errors.index(min(lin_errors))]   
plt.plot(alphas,lin_errors,'r-')
```

```python
# use the alpha that gave the best result to build the final model
best_lin_model = train_model(randomized_training_set, x_parameter_list, 'DomesticTotalGross', alphas[lin_errors.index(min(lin_errors))])

# calculate R^2
print "R-Squared: %.3f" % r2_score(randomized_testing_set['DomesticTotalGross'], best_lin_model.predict(randomized_testing_set[x_parameter_list]))

# Plot predictions versus actuals
plt.figure(figsize=[15,5])
plot_predictions(randomized_testing_set, x_parameter_list, 'DomesticTotalGross', best_lin_model)
plt.xlabel('Actual Domestic Gross')
plt.ylabel('Predicted Domestic Gross')
```

```python
# Plot training and testing errors to check for convergence
lin_training_errors = []
lin_testing_errors = []
for i in range(1,201):
    block = int(i) * 5
    test_training_set = randomized_training_set[0:block]
    test_testing_set = randomized_testing_set[0:block]
    lin_training_errors.append(mean_squared_error(test_training_set['DomesticTotalGross'],best_lin_model.predict(test_training_set[x_parameter_list])))
    lin_testing_errors.append(mean_squared_error(test_testing_set['DomesticTotalGross'],best_lin_model.predict(test_testing_set[x_parameter_list])))
x = [i for i in range(1,201)]

plt.figure(figsize=[15,5])
plt.plot(x,lin_training_errors,'r-')
plt.plot(x,lin_testing_errors,'b-')
plt.xlabel('# Data Points')
plt.ylabel('Mean Squared Error')
```
